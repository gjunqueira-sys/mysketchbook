# Signal Processing

## Real-Value Exponential 

```julia
r(a, n) = a^n
```

```julia
r_exp = [r(1.1, n) for n in -50:50]
```

```julia
using Plots
```

```julia
plot(r_exp, line=:stem, marker=:o)
```

For $a > 1$,  $r[n]$, the curve srhrinks to the left and increase to the right.

```julia
r_exp = [r(0.9, n) for n in -50:50]
```

```julia
plot(r_exp, line=:stem, marker=:o)
```

For  $0 < a < 1$ , $r[n]$ grows to the left and shrinks to the right.
