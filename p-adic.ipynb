## 1. p-adic Mod 3

```julia
using Statistics, Plots, PlutoUI, LinearAlgebra, Symbolics
```

```julia
N = 100; #CREATE A NATURAL NUMBER VECTOR
```

```julia
p3s= [mod(i,3) for i in 1:N]
```

```julia
function padic(X::Int64)
    return mod(X,3)
end
    
```

```julia
S = [i for i in 0:N if padic(i)==0];
@show S;
```

```julia
S = [i for i in 0:N if padic(i)==1];
@show S;
```
