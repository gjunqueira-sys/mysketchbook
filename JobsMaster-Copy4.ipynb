```python
#load csv file containing jobmaster excel sheet, parse based on SAP DATE Entry column
df = pandas.read_csv('jobsmaster.csv', sep=',',parse_dates=['SAP DATE'])
df=df[['SO','Customer', 'SAP DATE', 'SHIP DATE', 'REVENUE', 'ENGINEER', 'Estimated Eng. Hours','Number of Weeks']]#clean table
```

```python
def getEngineerJobs (eng, date, df):
    """function return jobs listed for specific engineer.
        @@params 
        eng = string name/code of engineer
        date = return engineer jobs beyond a certain SAP DATE, format yyyy-MM-dd ie '2017-06-30'
        df  = dataframe
        @@@return: dataframe filtered for specific engineer
        @@@@example of using function call
        DK_JOBS=getEngineerJobs ('DK',df)
    """
    Filter = ((df['ENGINEER']== eng) & (df['SAP DATE']> date))
    
    return df[Filter]
    
```

```python
def getEngineerEstimatedHours (dfeng):
    """function return estimated hours from a given engineer
    @@params
    dfeng: dataframe filtered for a specific engineer (via getEngineerJobs)
    @@@return:dataframe with dates and estimate hours
    @@@@example of using function call:
    DF4=getEngineerEstimatedHours (DK_JOBS)
    DF4.plot()
    plt.show()
    
    """
    DATES=dfeng[['SAP DATE','Estimated Eng. Hours']].groupby('SAP DATE')
    return DATES.sum()
```

```python
def getEngineerRevenue(dfeng):
    """function return revenue assigned to a given engineer
    @@params
    dfeng: dataframe filtered for a specific engineer (via getEngineerJobs)
    @@@return:tuple with dates and descriptive statistics of jobs
        [0] groupby dates and Sum of revenues
        [1] groupby dates and revenue mean
        [2] groupby dates and revenue min
        [3] groupby dates and revenue max
        [4] groupby dates and revenue count
        [5] total revenue
    @@@@example of using function call:
    DF4=getEngineerRevenue (DK_JOBS)
    DF4[0].plot()
    
    plt.show()
    
    """
    DATES=dfeng[['SAP DATE','REVENUE']].groupby('SAP DATE')
    return (DATES.sum(),DATES.mean(),DATES.min(),DATES.max(),DATES.count(),DATES.sum().sum())
```

```python
def getEngineerWorkload (dfeng):
    """function to return workload for a given engineer
    @@params
    dfeng dataframe for a particular engineer
   
    @@@returns
    [0] GroupBy object with Workloads and dates
    [1] Matrix structure for workload
    [2] Data structure for workload
    
    """
    
    MaxNumberOfRows = dfeng.shape[0]
    MaxNumberOfColumns = dfeng.shape[1]
    workloads=[]
    Capacity=[]
    dates=[]
    i=0
    
    #create matrix to hold new structure
    Matrix = np.zeros(shape=(2000, 60))
    MatrixCap=np.zeros(shape=(2000, 60))
    
    for index, row in dfeng.iterrows():
        t=np.linspace(1,int(row['Number of Weeks']),int(row['Number of Weeks']))
        rng = pandas.date_range(row['SAP DATE'], periods=int(row['Number of Weeks']), freq='W')
        
        for weeks in range (0,int(row['Number of Weeks'])):
            M=row['Estimated Eng. Hours']
            Ni=0.1
            c=1.0
            Matrix[index][weeks]=M/(1.0+((M-Ni)/Ni)*np.e**(-c*t[weeks]))
            MatrixCap[index][weeks]=40.0
            workloads.append((Matrix[index][weeks]))
            Capacity.append(MatrixCap[index][weeks])
            dates.append(rng[weeks])
            
        
            
    data = {'dates': dates, 'work': workloads}
    data2 = {'dates': dates, 'work': Capacity}
    
    df = pandas.DataFrame(data=data)
    
    dfwork = df.groupby('dates').sum()
    dfworkf = df.groupby('dates').sum()
    
    for i in range (0, dfwork['work'].shape[0]-1):
        
        if ( dfwork['work'][i+1] -  dfwork['work'][i])>0:
            dfworkf['work'][i+1]= ( dfwork['work'][i+1] -  dfwork['work'][i])
            
        else:
            dfworkf['work'][i+1]=0
                
   
   
    
    return dfworkf.groupby('dates').sum(), Matrix, data, data2, df
            
```
